Linear Regression model

```python
class Linear_Regression:
  #initiating the parameteers learing rate and no of
  def __init__(self, learning_rate, no_of_iterations):

    self.learning_rate= learning_rate
    self.no_of_iterations = no_of_iterations


  def fit(self,X,Y):

    #no of training examples and number of features
    self.m , self.n= X.shape# no of rows and columns

    # intiating the weight and bias

    self.w = np.zeros(self.n)
    self.b =0

    self.X = X
    self.Y = Y



    # implement the grdienr descent

    for i in range(self.no_of_iterations):
      self.update_weights()


  def update_weights(self,):

    Y_prediction = self.predict(self.X)


    #calculate gradients

    dw = -(2 * (self.X.T).dot(self.Y - Y_prediction)) / self.m
    db = -2 * np.sum(self.Y - Y_prediction)/ self.m

    # updating weights

    self.w = self.w - self.learing_rate*dw
    self.b = self.b- self.learning_rate*db



  def predict(self,X):

    return X.dot(self.w)+self.b


```
